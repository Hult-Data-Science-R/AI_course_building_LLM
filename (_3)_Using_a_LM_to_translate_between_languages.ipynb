```python
# we will use a facebook model to translate from english to a different language:
#https://huggingface.co/facebook/nllb-200-distilled-600M

checkpoint = 'facebook/nllb-200-distilled-600M'
# checkpoint = ‘facebook/nllb-200–1.3B’
# checkpoint = ‘facebook/nllb-200–3.3B’
# checkpoint = ‘facebook/nllb-200-distilled-1.3B’
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
# I want to translate from english to spanish ... I don't know spanish!! :)

source_lang = 'eng_Latn'
target_lang = 'spa_Latn'

# I can check other language repos here: https://github.com/facebookresearch/flores/blob/main/flores200/README.md#languages-in-flores-200
```

```python
translator = pipeline('translation', model=model, tokenizer=tokenizer, src_lang=source_lang, tgt_lang=target_lang, max_length = 400)
```

```python
text = 'I want to translate from english to spanish ... I don\'t know spanish!!'

output = translator(text)
translated_text = output[0]['translation_text']
print(translated_text)
```

```python
# now trying to convert english to polish - who knows polish?

source_lang = 'eng_Latn'
target_lang = 'pol_Latn'
translator = pipeline('translation', model=model, tokenizer=tokenizer, src_lang=source_lang, tgt_lang=target_lang, max_length = 400)

text = 'I want to translate from english to spanish ... I don\'t know spanish!!'

output = translator(text)
translated_text = output[0]['translation_text']
print(translated_text)
```
