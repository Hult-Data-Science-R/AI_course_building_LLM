```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# Testing a different model type from Hugging Face

result = classifier("I love using Hugging Face models in Colab!")
print(result)
```

```python
# Trying a totally different sentiment model :
# https://huggingface.co/tabularisai/multilingual-sentiment-analysis
#this one offers multilingual sentiment

classifier = pipeline("sentiment-analysis", model="tabularisai/multilingual-sentiment-analysis")

result = classifier("I love using Hugging Face models in Colab!")
print(result)
```

```python
# Using the same model as above:
# https://huggingface.co/tabularisai/multilingual-sentiment-analysis
# try sentiment for different test cases - including different languages:

#Example number 1

result = classifier("I really hate data science and cannot learn how to code in Python ... my worst nighmare")
print(result)
```

```python
# Using the same model as above:
# https://huggingface.co/tabularisai/multilingual-sentiment-analysis
# try sentiment for different test cases - including different languages:

#Example number 2

result = classifier("Bardzo lubie kwaitki i Warszawa. Kiedy jestem w Polsce to zawsze jade nad morze zjesc gofry ktore sa pyszne.")
print(result)
```
