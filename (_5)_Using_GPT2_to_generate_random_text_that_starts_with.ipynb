```python
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model='gpt2')

```

```python
# using the GPT2 model to generate random text that would follow a prompt:
set_seed(42)
generator("Hello, I'm a language model,", max_length=30, num_return_sequences=3)
```

```python
# testing for another example
generator("I'm a Master's level student in San Francisco, and I should focus on studying ", max_length=30, num_return_sequences=3)
```
